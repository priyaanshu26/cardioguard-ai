```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df.drop(['id'], axis=1)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df['age_years'] = (df['age'] / 365).astype(int)
```

```python
df = df.drop(['age'], axis=1)
```

```python
df.describe()
```

```python
sns.histplot(df['age_years'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.histplot(df['height'], kde=True)
plt.title('Height Distribution')
plt.show()
```

```python
sns.histplot(df['weight'], kde=True)
plt.title('Weight Distribution')
plt.show()
```

```python
df[['ap_hi', 'ap_lo']].skew()
```

```python
#Ranges of all the parameters
print('Height range:', df['height'].min(), "-", df['height'].max())
print('Weight range:', df['weight'].min(), "-", df['weight'].max())
print('ap_hi range:', df['ap_hi'].min(), "-", df['ap_hi'].max())
print('ap_lo range:', df['ap_lo'].min(), "-", df['ap_lo'].max())
```

```python
df['cardio'].value_counts()
```

```python
#Normal height
df = df[(df['height'] >= 120) & (df['height'] <= 215)]
```

```python
#Normal weight
df = df[(df['weight'] >= 40) & (df['weight'] <= 180)]
```

```python
#Average BP
df = df[(df['ap_hi'] >= 60) & (df['ap_hi'] <= 240)]
df = df[(df['ap_lo'] >= 30) & (df['ap_lo'] <= 150)]
df = df[df['ap_hi'] > df['ap_lo']]
```

```python
df.shape
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df.corr(), kde=False)
plt.title('Cardio Disease Distribution')
plt.show()
```

```python
sns.boxplot(x='cardio', y='age_years', data=df)
plt.title('Age vs Cardio')
plt.show()
```

```python
sns.boxplot(x='cardio', y='ap_hi', data=df)
plt.title('Systolic BP vs Cardio')
plt.show()
```

```python
df['BMI'] = df['weight'] / ((df['height']/100) ** 2)

sns.boxplot(x='cardio', y='BMI', data=df)
plt.title('BMI vs Cardio')
plt.show()
```

```python
lifestyle = ['cholesterol', 'gluc', 'smoke', 'alco', 'active']

for col in lifestyle:
    sns.countplot(x=col, hue='cardio', data=df)
    plt.title(f'{col} vs Cardio')
    plt.show()
```

```python
df.to_csv("cardio_clean.csv", index = False)
```
